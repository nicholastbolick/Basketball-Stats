```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
on_court = pd.read_csv('cbb.csv')
on_court['Win Avg'] = on_court['W'] / on_court['G']
first_column = on_court.pop('Win Avg')
on_court.insert(1, 'Win Avg', first_column)
on_court.pop('G')
on_court.pop('W')
on_court.pop('SEED')
on_court.pop('YEAR')
on_court.pop('WAB')
on_court.pop('POSTSEASON')
on_court.pop('ADJOE')
on_court.pop('ADJDE')
on_court.pop('BARTHAG')
col_names = []
for col in on_court.columns:
    col_names.append(col)
    
print(col_names)

correlation_df = on_court.corr().sort_values('Win Avg',ascending=False)
correlation_df.drop(columns='Win Avg')
correlation_df
```
